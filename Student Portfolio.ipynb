La maggior parte degli studenti dedica allo studio dalle 2 alle 5 ore settimanali indipendentemente dalla durata del viaggio. La durata del tragitto casa-scuola non sembra avere un impatto molto forte sul tempo settimanale dedicato allo studio, tuttavia non possiamo avere una rappresentazione precisa. Infatti sebbene le due varaibili abbiano lo stesso range di valori e siano tutte e due categoriche ordinali,  una rappresenta la durata **giornaliera** del viaggio e l'altra il tempo **settimanale** dedicato allo studio. Avendo a che fare con dei range di valori e non con dei numeri effettivi non posso convertire il tempo settimanale dedicato allo studio in tempo giornaliero dedicato allo studio; inoltre bisognerebbe tener conto del fatto che gli studenti vanno ascuola soltanto per 5 o 6 giorni alla settimana mentre sono liberi di studiare anche 7 giorni alla settimana. Non è possibile quindi stabilire quale sia il reale impatto della durata del viaggio rispetto al tempo dedicato allo studio; se avessimo a disposizione per ogni studente il tempo dedicato al viaggio e allo studio in ore o minuti potremo facilmente calcolare e visualizzare questa relazione. Possiamo però visualizzare in maniera più accurata come si distribuisce il tempo dedicato allo studio quando il viaggio dura 30 o più minuti:

```python
#Considero soltanto la durata del viaggio maggiore uguale a 30 minuti
above_30 = df[df['traveltime']>2]

sns.countplot(x= 'traveltime', hue='studytime', data=above_30)
plt.xticks(np.arange(2),['30 min - 1 ora', '> 1 ora']) 
plt.title('Tempo di studio settimanale Vs Durata giornaliera viaggio \n maggiore di 30 minuti')
plt.xlabel('Durata del viaggio')
plt.legend(labels=['<2h','2-5h','5-10h','>10h'], title='ore settimanali \n dedicate allo studio');
```

Possiamo calcolare la correlazione fra il tempo dedicato allo studio e la durata del viaggio utilizzando il coefficiente tau di Kendall, in quanto le due varaibili sono categoriche ordinali e non numeriche; sono appaiate sugli stessi casi ed esiste tra le due una relazione monotona. 

```python
corr = df[['traveltime','studytime']].corr(method='kendall')
matrix=np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True,mask=matrix, cbar_kws={'label': 'Correlazione'},square=True)
plt.title('Correlazione fra tempo dedicato allo studio \n e durata del viaggio');
```

Dalla matrice di correlazione possiamo vedere che le due variabili hanno infatti una correlazione negativa pari a -0.4. Ciò vuol dire che all'aumentare della durata del viaggio corrisponde una riduzione del tempo dedicato allo studio.

# Machine Learning

Per questo dataset addestro 3 diversi modelli: **RandomForest**, **SupportVectorRegression** e **KNeighborsRegressor**. Per validare i gli iperparametri di ogni modello utilizzerò **RandomizedSearch**.

## RandomizedSearchCV

RandomizedSerchCV è un metodo di validazione degli iperparametri che funziona in modo diverso rispetto a GridSearchCV. Con il metodo RandomizedSearchCV possiamo definire una griglia o una distribuzione di valori per gli iperparametri da validare per poi campionare delle combinazioni casuali di questi iperparametri, su cui viene effettuata una cross validation, fino a trovare la combinazione ideale. Con questo approccio otteniamo sostanzialmente due vantaggi:

- Si può scegliere un budget (con budget si intende tempo e potenza di calcolo) indipendentemente dal numero di parametri e dai possibili valori assegnati.

- Aggiungere altri parametri che non influenzano la performance del modello, non ne riduce l'efficienza.

Per specificare da dove devono essere campionati i paramerti si usa un dizionario come con GridSearchCV. Il parametro *n_iter* è quello che controlla il budget computazionale.Con questo parametro modifichiamo il numero dei valori candidati al campionamento o delle iterazioni del campionameto. 

Valuto la performance degli modelli sul train test con l'R2 per avere un'idea generale delle performance del modello, una volta effettuate le previsioni valuto i modelli utilizzando anche altre metriche.

## RandomForest

Il parametro più importante di random forest è ***n_estimators*** che controlla il numero di alberi nella foresta, cioè il numero di decision tree che compongono il modello. di solito più il numero è elevato e più il modello è performante ma ciò si traduce anche in un tempo di calcolo più lungo. Inoltre è importante notare che i risultati non miglioreranno in modo significativo dopo un certo numero di alberi. ***max_features*** invece è un iperparametro e indica la dimensione del campione casuale di features da prendere in considerazione quando si divide un nodo. Più è basso questo numero e maggiore sarà la riduzione della varianza. Un altro iperparametro che può essere utile validare è ***min_samples_leaf*** che indica il numero minimo di campioni contenuti in una foglia per poter effettuare uno split, nelle regressioni può avere un effetto di smoothing. 

Cominciamo defindendo i parametri di random forest che vogliamo validare utilizzando RandomizedSearch.

```python
#numero di alberi
n_estimators = [150,200,300,400,500]

#Numero di feature da considerare per lo split
max_features = [int(x) for x in np.linspace(start=1, stop=X_train.shape[1], num=40)]

#Numero minimo di campioni richiesti ad ogni nodo
min_samples_leaf =loguniform(1e-3,1e-1) 

#Creo un dizionario con i parametri
random_grid = {'n_estimators': n_estimators,
              'max_features': max_features,
              'min_samples_leaf': min_samples_leaf}

#Creo il modello base da validare
rf = RandomForestRegressor(random_state=42)
#RandomizedSearch
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,
                               n_iter=200, scoring='r2', 
                               cv=5, random_state=42, n_jobs=-1)

#Fit the model
rf_random.fit(X_train, y_train);
```

```python
print('migliori parametri:{}'.format(rf_random.best_params_))
print('score: R2 validazione:{}'.format(rf_random.best_score_))
```

## Support Vector Regression

Le Support Vector Machines possono essere utilizzate anche per problemi di regressione con Support Vector Regressor. SVR ha come obiettivo quello di minimizzare l'errore sul training set. SVR però funziona in modo diverso rispetto a SVC. Invece di cercare il margine più ampio fra due classi limitando al contempo le violazioni del margine, il regressore cerca di fare rientrare più istanze possibili all'interno del margine sempre cercando di limitare le violazioni. La funzione obiettivo di SVR minimizza i coefficienti - in particolare la l2-norm del vettore dei coefficienti - e non l'errore quadratico. Il termine di errore quindi viene gestito nei vincoli, dove possiamo imporre che l'errore assoluto sia pari o minore ad una determinata soglia definita come _errore massimo_ (epsilon); mentre la penalità di violazione dei vincoli viene gestita da C. 
Il parametro **epsilon** può aiutarci a controllare l'accuratezza del nostro modello. Questo parametro infatti definisce un margine di errore all'interno del quale non viene assegnata penalità agli errori.
Un altro iperparametro da considerare durante la costruzione di un regressore SVR è **C** il cui obiettivo è pesare la violazione dei vincoli. Più è elevato e più pagherò la violazione dei punti del training set per cui sbaglio più di epsilon. 
Infine inseriamo anche varie opzioni per il **kernel**. Il kernel è una funzione che prende come input due vettori e restituisce il loro prodotto scalare in un nuovo spazio delle features con una dimensionalità più elevata rispetto allo spazio di partenza. Le funzioni vengono appunto usate come parametri in SVM e ci aiutano a determinare la forma dell'iperpiano e il confine decisionale. Intuitivamente quando abbiamo un dataset che non è separabile data l'attuale dimensionalità, possiamo aggiungere e manipolare ulteriori dimensioni grazie ai kernel. Consideriamo ad esempio un dataset 2D in cui i nostri dati non sono linearmente separaibili.



Tuttavia questi dati potrebbero essere interpolabili da un iperpiano se li trasportassimo in uno spazio vettoriale 3D. Per effettuare questa trasformazione utilizziamo una funzione Φ(). L'esempio dell'immagine è piuttosto semplice ed intuitivo tuttavia quando abbiamo a che fare con molte dimensioni il costo computazionale e la complessità crescono perché dovremo calcolare Φ(x) per ognuna delle nostre x. In questi casi possiamo ricorrere al kernel **trick** che ci permette di operare nello spazio delle feature originale senza calcolare le coordinate dei dati in un spazio con dimensionalità più elevata.

Ogni kernel per le SVM dispone anche di alcuni iperparametri peculiari per permetterci di regolare al meglio il modello al fine di contrllare efficacemente il trade-off fra bias e varianza.

```python
#Definizione parametri

epsilon = loguniform(1e0, 1e3)

C = loguniform(1e-3, 1e3)

kernel = ['rbf']

gamma = loguniform(1e-4, 1e-3)

random_grid = {'epsilon': epsilon,
              'C': C,
              'kernel':kernel,
              'gamma':gamma}

#costruzione modello
svr = SVR() 

svr_random = RandomizedSearchCV(estimator = svr, param_distributions = random_grid,
                               n_iter=300, scoring='r2', 
                               cv=5, random_state=42, n_jobs=-1)
svr_random.fit(X_train, y_train);
```

```python
print('migliori parametri:{}'.format(svr_random.best_params_))
print('score: R2 validazione:{}'.format(svr_random.best_score_))
```

## KNeighborsRegressor

L'algoritmo K Neighbors può essere usato anche per le regressioni utilizzando la stessa logica del classificatore. Anche il regressore si basa sui valori 'vicini' per calcolare il nuovo valore per la previsione. KNeighborsRegressor è in grado di gestire molto bene la non linearità dei dati. 
Come per il calssificatore il parametro più importante è ***n_neighbors***, cioè il parametro _K_. Un valore di K elevato contribuisce a ridurre il rischio di overfit generando un confine decisionale meno frastagliato ma a spese di una maggiore varianza. Un altro iperparametro da validare è ***weights***. Questo parametro serve a controllare il peso atribuito ai 'vicini'. Può avere due valori:
- 'uniform': tutti i punti hanno lo stesso valore
- 'distance': i punti vicini hanno un'influenza maggiore rispetto ai punti più lontani.

```python
#Definizione parametri
n_neighbors = np.arange(1,10)
weights = ['uniform', 'distance']

random_grid = {'n_neighbors': n_neighbors,
              'weights': weights}

neigh = KNeighborsRegressor()
neigh_random = RandomizedSearchCV(estimator = neigh, param_distributions = random_grid,
                               n_iter=20, scoring='r2', 
                               cv=5, random_state=42, n_jobs=-1)

neigh_random.fit(X_train, y_train);
```

```python
print('migliori parametri:{}'.format(neigh_random.best_params_))
print('score: R2 validazione:{}'.format(neigh_random.best_score_))
```

RandomForest e SupportVectorRegression hanno delle performance simili mentre KneighborRegressor è quello con la performance peggiore. La metrica usata per la valutazione è l'R². Questa metrica indica la performance generale del modello e rappresenta la quota di varianza (di y) che è stata spiegata dalle variabili indipendenti (le nostre features). Ci fornisce una misura della "goodness of fit" cioè una misura di quanto sia probabile che il modello effettui previsioni corrette su dati mai visti. Il valore migliore possibile è 1, difficilmente ottenibile nella realtà, quindi modelli con un R² vicino ad 1 dovrebbero essere in grado di fornire buone previsioni anche su dati mai visti prima.

# Previsioni e valutazione

A questo punto possiamo usare i modelli addestrati per efefttuare previsioni su dati mai visti prima. Dopo le previsioni analizzeremo le loro performance.

```python
#Funzione per calcolare le previsioni e plottare il grafico degli errori

def score(model):
    print(model.best_estimator_)
    print('-'*50)
    preds=model.predict(X_test)
    
    print("RMSE:",np.sqrt(mean_squared_error(y_test,preds)))
    print('r2_score:',r2_score(y_test,preds))
    print('MAE:', mean_absolute_error(y_test,preds))
    
    fig, ax = plt.subplots(figsize=(8,8))
    ax.scatter(x=y_test,y=preds,color='#E76F51')
    ax.plot([0, 1], [0, 1], color='black', transform=ax.transAxes)
    plt.title('Valore Reale Vs Previsioni',fontsize=14)
    plt.xlabel('Valori Reali',fontsize=14)
    plt.ylabel('Previsioni',fontsize=14)
```

```python
#RandomForestRegressor
score(rf_random)
```

```python
#SVR
score(svr_random)
```

```python
#KNeighborsRegressor
score(neigh_random)
```

Anche sui dati mai visti prima le performance dei tre modelli restano piuttosto simili ma il modello migliore dei tre è RandomForestRegressor. Graficamente non è possibile notare una grande differenza nelle performace dei modelli ma osservando le metriche di valutazione si può vedere che RandomForest è sia il modello con R2 più elevato che con RMSE e MAE più bassi. 

Abbiamo già parlato dell'R² mentre gli altri due indicatori di performance sono il ***Root mean squared error (RMSE)*** e il ***mean absolute error (MAE)***  A differenza dell'R², queste due metriche di valutazione sono _loss function_ (funzioni di costo) che puntiamo a minimizzare, quindi valori più piccoli indicano una migliore performance.

Il RMSE è la radice quadrata del mean squared error (MSE) che è definito come la differenza al quadrato fra valore effettivo e valore predetto. Praticamente MSE rappresenta la distanza al quadrato fra valori reali e valori predetti. RMAE non è robusto rispetto ai valori outlier ( se abbiamo valori estremi nel dataset penalizza gli outlier e il MSE risulta più grande a causa dell'elevamento al quadrato). Di solito si usa la radice quadrata di questa metrica perché è più facile da confrontare visto che ha la stessa unità di misura della variabile target, mentre MSE avrebbe l'unità di misura elevata al quadrato.

MAE calcola la differenza in valore assoluto dei valori reali e dei valori previsti. Ho affiancato ad RMSE anche MAE perché non è influenzato dai valori outlier in modo da ottenere una valutazione più completa del modello.

A differenza dell'R² sia per RMSE che per MAE valori minori indicano perfomance migliori del modello perché entrambi misurano quanto le previsioni si avvicinano ai valori reali.

Per calcolare l'importanza delle feature possiamo usare anche permutation_importance e vedere come gli altri modelli hanno trattato le diverse feature.

```python
from sklearn.inspection import permutation_importance
r_1 = permutation_importance(rf_random, X_test, y_test,
                           n_repeats=40,
                           random_state=42)
for i in r_1.importances_mean.argsort()[::-1]:
     if r_1.importances_mean[i] - 2 * r_1.importances_std[i] > 0:
         print(f"{feature_names[i]:<8}"
               f"{r_1.importances_mean[i]:.3f}"
               f" +/- {r_1.importances_std[i]:.3f}")
```

```python
r_2 = permutation_importance(svr_random, X_test, y_test,
                           n_repeats=40,
                           random_state=42)
for i in r_2.importances_mean.argsort()[::-1]:
     if r_2.importances_mean[i] - 2 * r_2.importances_std[i] > 0:
         print(f"{feature_names[i]:<8}"
               f"{r_2.importances_mean[i]:.3f}"
               f" +/- {r_2.importances_std[i]:.3f}")
```

```python
r_3 = permutation_importance(neigh_random, X_test, y_test,
                           n_repeats=40,
                           random_state=42)
for i in r_3.importances_mean.argsort()[::-1]:
     if r_3.importances_mean[i] - 2 * r_3.importances_std[i] > 0:
         print(f"{feature_names[i]:<8}"
               f"{r_3.importances_mean[i]:.3f}"
               f" +/- {r_3.importances_std[i]:.3f}")
```
